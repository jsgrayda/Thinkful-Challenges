```python
raw_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/epi_r.csv')
```

```python
column_names = list(raw_data.columns)

for eachColumn in column_names:
    raw_data[eachColumn].fillna(raw_data[eachColumn].median(), inplace = True) # replace empty spot with median
```

```python
from scipy.stats.mstats import winsorize
from sklearn.preprocessing import scale
```

```python
X = raw_data.iloc[:, : 30] # drop('rating', axis=1).
y = round(raw_data['rating'])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```
